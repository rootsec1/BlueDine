```python
import json
import os
from pocketbase import PocketBase

client = PocketBase("http://127.0.0.1:8090")
user_data = client.collection("users").auth_with_password(
    "abhishek.murthy@duke.edu",
    "password"
)
print(user_data.record)

files = os.listdir("../data/raw")
for file in files:
    if not file.endswith("_summary.json"):
        continue

    with open(f"../data/raw/{file}") as f:
        item_list = list(json.load(f))
        for item in item_list:
            pb_item = {
                "name": item["name"],
                "description": item.get("description", ""),
                "price": item["price"],
                "type": item["type"],
                "ingredients": ";".join(item["ingredients"]),
                "calories": item["calories"],
                "category": item["category"],
                "restaurant": item["restaurant"],
                "time_open": item["timeopen"],
                "time_close": item["timeclose"],
            }
            pb_response = client.collection("items").create(pb_item)
            print(pb_response)
```
